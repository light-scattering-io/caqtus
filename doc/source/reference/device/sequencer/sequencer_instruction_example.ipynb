For example, a complex instruction can be created as follow:

```python
from caqtus.device.sequencer.instructions import (
    Pattern,
    ramp,
    plot_instruction,
)

instr = (
    (Pattern([0]) * 2 + Pattern([1])) * 6
    + Pattern([0]) * 5
    + ramp(0.0, 1.5, 20)
    + Pattern([1.5]) * 15
)

instr
```

It represents the following sequence of values:

```python
plot_instruction(instr);
```
